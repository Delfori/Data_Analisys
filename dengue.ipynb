```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

читаем с txt и грузим в датафрейм

```python
file = os.path.join('datasets','dengue','ph_dengue_cases2016-2020.csv')
```

```python
df = pd.read_csv(file)
df.head()
```

```python
# Преобразуем колонки года и месяца в одну вида datetime и назначим её индексом
df.index = pd.to_datetime(df['Year'].astype(str)  + df['Month'], format='%Y%B')
# уберём ненужные колонки через слайсинг
df = df.loc[:, 'Region':'Dengue_Deaths']
```

```python
df.plot()
```

```python
df.groupby(df.index.month).sum().plot(kind='bar')
```

```python
subdf = df[df.index.year.isin([2019,2020])].copy()
subdf.groupby(subdf.index.month).sum().plot(kind='bar')
```
